**1).** Сколько строк с мужчинами из категории города A? (речь не об уникальных ID мужчин, а о количестве строк)  

```python
print(data[(data['Gender'] == 'M') & (data['City_Category'] == 'A')].describe()['User_ID']['count'])
print(data[data['Gender'] == 'M']['City_Category'].value_counts())
data[(data['Gender'] == 'M') & (data['City_Category'] == 'A')].describe()
```

**2).** Сколько женщин от 46 до 50, потративших (столбец Purchase) больше 20000 (условных единиц, в данном случае)?   (речь не об уникальных ID, а о количестве строк)  

```python
print(data[(data['Gender'] == 'F') & (data['Age'] == '46-50') & (data['Purchase'] > 2000.0)].describe()['User_ID']['count'])
print(data[(data['Gender'] == 'F') & (data['Purchase'] > 2000.0)]['Age'].value_counts())
data[(data['Gender'] == 'F') & (data['Age'] == '46-50') & (data['Purchase'] > 2000.0)].describe()
```

**3).** Сколько NaN'ов в столбце Product_Category_3?  

```python

print(data['Product_Category_3'].unique() )
print(data.count()['User_ID'] - data['Product_Category_3'].count(), 'Столько нанов')
print(data['Product_Category_3'].value_counts())
```

**4).** Какую долю (вещественное число от 0 до 1, округлить до 4-го знака) от всех покупателей составляют ВМЕСТЕ мужчины от 26 до 35 лет и женщины старше 36 лет (то есть нужно учесть несколько возрастных категорий)? (речь не об уникальных ID, а о количестве таких строк)  

```python
print(data[
    (data['Gender'] == 'M') & (data['Age'] == '26-35')
].describe())

print(data[(data['Gender'] == 'F') & ( np.any(data['Age'] == ['55+', '46-50', '51-55', '36-45']))])
```
